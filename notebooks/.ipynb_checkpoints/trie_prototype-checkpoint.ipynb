### Example: Small Sample Dataset

```python
career_list = [
    "Data Scientist",
    "Software Engineer",
    "Project Manager",
    "Teacher",
    "Banker",
    # ... eventually full O*NET dataset
]
```

```python
# Initialize Trie
career_trie = Trie()

# Insert careers into Trie
for career in career_list:
    career_trie.insert(career)
```

### Test autocomplete for some prefixes

```python
prefix1 = "Da"
prefix2 = "S"

# Call trie autocomplete (returns list of matching careers)
results1 = career_trie.autocomplete(prefix1)  # expected: ["Data Scientist"]
results2 = career_trie.autocomplete(prefix2)  # expected: ["Software Engineer"]

# Display results
print("Results for", prefix1, ":", results1)
print("Results for", prefix2, ":", results2)
```
